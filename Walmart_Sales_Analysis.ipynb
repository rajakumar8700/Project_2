***according to calender***
- 1st quarter : jan to march
- 2nd quarter : april to jun
- 3rd quarter : july to sep

```python
q2_2012 = wm[(wm['Date'] >= '2012-04-01') & (wm['Date'] <= '2012-06-30')]
q3_2012 = wm[(wm['Date'] >= '2012-07-01') & (wm['Date'] <= '2012-09-30')]

q2_Sales = q2_2012.groupby('Store')['Weekly_Sales'].sum().rename('Q2_sales_result')
q3_Sales = q3_2012.groupby('Store')['Weekly_Sales'].sum().rename('Q3_sales_result')

# combining q2 and q3 sales
quarterly_sales = pd.concat([q2_Sales,q3_Sales],axis=1).dropna()
quarterly_sales['Growth_rate'] = (
    (quarterly_sales['Q3_sales_result'] - quarterly_sales['Q2_sales_result'])/(quarterly_sales['Q2_sales_result'])
)
quarterly_sales['Growth_rate %'] = ((quarterly_sales['Growth_rate'])*100).round(2)
top_growth_stores = quarterly_sales.sort_values(by='Growth_rate',ascending=False)
top_growth_stores
```

***we can say that***
- *Store 7 had the highest growth rate at +13.33% followed by Store 16 (+8.49%), Store 35 (+4.47%), and so on.*

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_stores = top_growth_stores.head(10)
plt.figure(figsize=(18,6))
sns.barplot(data=top_stores,y='Growth_rate',x=top_stores.index,palette='viridis',edgecolor='black')

plt.title('Store-wise Q3 2012 Growth Rate')
plt.xlabel('Store')
plt.ylabel('Growth Rate')
plt.tight_layout()
plt.show()
```

```python
# Convert 'Date' column to datetime if not already
wm['Date'] = pd.to_datetime(wm['Date'])

# Define holiday event dates
holiday_events = {
    'Holiday': ['Super Bowl', 'Super Bowl', 'Super Bowl', 'Super Bowl',
                'Labour Day', 'Labour Day', 'Labour Day', 'Labour Day',
                'Thanksgiving', 'Thanksgiving', 'Thanksgiving', 'Thanksgiving',
                'Christmas', 'Christmas', 'Christmas', 'Christmas'],
    'Date': ['2010-02-12', '2011-02-11', '2012-02-10', '2013-02-08',
             '2010-09-10', '2011-09-09', '2012-09-07', '2013-09-06',
             '2010-11-26', '2011-11-25', '2012-11-23', '2013-11-29',
             '2010-12-31', '2011-12-30', '2012-12-28', '2013-12-27']
}

# Create a DataFrame of holidays
holidays_df = pd.DataFrame(holiday_events)
holidays_df['Date'] = pd.to_datetime(holidays_df['Date'])
if 'Holiday' in wm.columns:
    wm.drop(columns='Holiday', inplace=True)

# Merge holiday labels into main data
wm = wm.merge(holidays_df, how='left', on='Date')
wm['Holiday'] = wm['Holiday'].fillna('None')

# Calculate mean sales for non-holiday dates
non_holiday_mean = wm[wm['Holiday'] == 'None']['Weekly_Sales'].mean()

# Group by holiday and calculate mean sales
holiday_mean = wm[wm['Holiday'] != 'None'].groupby('Holiday')['Weekly_Sales'].mean()

# Filter holidays with higher mean sales than non-holiday average
higher_than_avg_holidays = holiday_mean[holiday_mean > non_holiday_mean]
print(higher_than_avg_holidays)
```

## **prediction analysis for store1**

```python
store1 = wm[wm['Store'] == 1].copy()
store1 = store1.sort_values('Date')
store1['Day'] = (store1['Date'] - store1['Date'].min()).dt.days + 1
```

```python
store1
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

X = store1[['Day', 'Fuel_Price', 'CPI', 'Unemployment']]
y = store1['Weekly_Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model = LinearRegression()
model.fit(X_train, y_train)

r2 = r2_score(y_test, model.predict(X_test))
print("RÂ² Score:", r2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = model.predict(X_test)

plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # Perfect prediction line
plt.xlabel('Actual Weekly Sales')
plt.ylabel('Predicted Weekly Sales')
plt.title('Actual vs Predicted Weekly Sales (Store 1)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Weekly Sales Trends Over Time

```python
# Quarterly Sales
quarterly_sales = wm.groupby('Quarter')['Weekly_Sales'].sum().reset_index()
quarterly_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,6))
sns.lineplot(x='Date', y='Weekly_Sales', data=wm, label='Weekly Sales')
plt.title('Weekly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
